```python
!pip install yfinance==0.1.67
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
Tesla=yf.Ticker("AAPl")
Tesla_info=Tesla.info
Tesla_share_data=Tesla.history(period="5y")
Tesla_share_data.reset_index(inplace=True)
k=Tesla_share_data["Close"]
b=Tesla_share_data["Date"]
plt.plot(b,k)
plt.xlabel("date")
plt.ylabel("price in dollars ($)")
plt.title("chart price of TESLA")
```
